```python
# Large language models couldn't produce word count limited text.
# so, if the final summaries of gpt4 is more than the given word count
# we send it back into the model to produce text within the given limit.
```

```python
import openai
```

```python
def chat(Messages):
    response = openai.ChatCompletion.create(
    model="gpt-4-1106-preview",
    messages=Messages
    )
    return response.choices[0].message.content
```

```python
# llama2 divide and conquer summaries / depth first- shorten
import os

def shorten_gpt4_summaries(input_folder, output_folder):
    count = 0
    for filename in os.listdir(input_folder):
         if filename.endswith(".txt"):
            print(filename)
            input_file_path = os.path.join(input_folder, filename)
            output_file_path = os.path.join(output_folder, filename)

            # read the content of the file
            with open(input_file_path, 'r', encoding='utf-8') as file:
                file_content = file.read()
                
            word_count = len(file_content.split())
            if word_count > 90:
                print('word count exceeded')
                count = count+1
                # method_1
                #new_messages = [
                        #{"role": "assistant", "content": file_content},
                        #{"role": "user", "content": "The summary that you provided exceeds the specified word limit. Please provide summary of the debate in 90 words without any opening statements. Ensure no assumptions are made and adhere strictly to the provided information." }
                    #] 
                # method_1 results are totally inappropiate
                # method_2
                new_messages = [
                        {"role": "system", "content": "As a text summarizer, you were tasked with summarizing a debate in 90 words. However, the summary you supplied exceeds the prescribed word limit. Please condense the provided summary to 90 words without making assumptions and strictly adhering to the given information."},
                        {"role": "user", "content": file_content }
                    ] 
                
                #print('Messages', Messages)
                final_summary = chat(new_messages)
                
                # write the concatenated result to a text file in the output folder  
                with open(output_file_path, 'w', encoding="utf8") as output_file:
                    output_file.write(final_summary)
            else:
                # within the limit, write the same content
                with open(output_file_path, 'w', encoding="utf8") as output_file:
                    output_file.write(file_content) 
    print('count',count)
```

```python
# input_folder_path - final summary produced by divide conquer/depth first approach
# output_folder_path - shortened summary

input_folder_path = "C:/Users/Nila/Documents/Main Scripts_25.12.2023/Jupyter Notebooks/complete testing/llm_output/GPT4/divide_conquer"
output_folder_path = "C:/Users/Nila/Documents/Main Scripts_25.12.2023/Jupyter Notebooks/complete testing/llm_output_updated/GPT-4/divide_conquer"

shorten_gpt4_summaries(input_folder_path, output_folder_path)
```
